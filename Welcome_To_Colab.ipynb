```
import pandas as pd
titanic_data={
              'PassengerId':[1,2,3,4,5],
              'Name':['a','b','c','d','e'],
              'Age':[10,11,12,13,14],
              'Sex':['m','f','m','f','m'],
              'Pclass':[3,1,3,1,3],
              'Survived':[0,1,1,1,0],
              'Fare':[7.25,71.2833,7.925,53.1,8.05],
              'Embarked':['S','C','S','S','S']
              }
df=pd.DataFrame(titanic_data)
df
```

```
print(df['Sex'].unique())
```

```
df['Sex']=df['Sex'].map({'m':0,'f':1})
df['Sex'].head()
```

```
df_encoded=pd.get_dummies(df,columns=['Embarked'],drop_first=True,dtype=int)
print(df_encoded.head())
```
